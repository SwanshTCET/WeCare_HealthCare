```python
import pandas as pd
c=pd.read_csv('dataset/Indian_Cities_Database.csv')
c
```

```python
cc= c[['State','Lat','Long']]
cc
```

```python
lat_long=[]
visited = []
for i in cc['State']:
        if(i not in visited):
            b = cc[cc['State']==i]
            latc = b['Lat'].mean()
            longc = b['Long'].mean()
            # print(i,latc,longc)
            visited.append(i)
            dict = {i:{'Lat':latc,'Long': longc}
            }
            lat_long.append(dict)
lat_long
           

    

```
